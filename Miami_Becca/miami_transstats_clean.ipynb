# Miami July Flights (2013-2018)

```python
# Read in csv files
flights_FL_201807 = pd.read_csv('BTS_data/FL_201807.csv')
flights_FL_201707 = pd.read_csv('BTS_data/FL_201707.csv')
flights_FL_201607 = pd.read_csv('BTS_data/FL_201607.csv')
flights_FL_201507 = pd.read_csv('BTS_data/FL_201507.csv')
flights_FL_201407 = pd.read_csv('BTS_data/FL_201407.csv')
flights_FL_201307 = pd.read_csv('BTS_data/FL_201307.csv')
```

```python
flights_FL_201807.shape
flights_FL_201807.head()
```

```python
flights_FL_201807.columns
```

```python
# subset flights that are from/to Miami from/to LA or New York
# flights from Miami
MIA_NYC_LA_201807 = flights_FL_201807[(flights_FL_201807['ORIGIN_CITY_NAME'] == 'Miami, FL') & ((flights_FL_201807['DEST_CITY_NAME'] == 'Los Angeles, CA') | (flights_FL_201807['DEST_CITY_NAME'] == 'New York, NY'))]
MIA_NYC_LA_201707 = flights_FL_201707[(flights_FL_201707['ORIGIN_CITY_NAME'] == 'Miami, FL') & ((flights_FL_201707['DEST_CITY_NAME'] == 'Los Angeles, CA') | (flights_FL_201707['DEST_CITY_NAME'] == 'New York, NY'))]
MIA_NYC_LA_201607 = flights_FL_201607[(flights_FL_201607['ORIGIN_CITY_NAME'] == 'Miami, FL') & ((flights_FL_201607['DEST_CITY_NAME'] == 'Los Angeles, CA') | (flights_FL_201607['DEST_CITY_NAME'] == 'New York, NY'))]
MIA_NYC_LA_201507 = flights_FL_201507[(flights_FL_201507['ORIGIN_CITY_NAME'] == 'Miami, FL') & ((flights_FL_201507['DEST_CITY_NAME'] == 'Los Angeles, CA') | (flights_FL_201507['DEST_CITY_NAME'] == 'New York, NY'))]
MIA_NYC_LA_201407 = flights_FL_201407[(flights_FL_201407['ORIGIN_CITY_NAME'] == 'Miami, FL') & ((flights_FL_201407['DEST_CITY_NAME'] == 'Los Angeles, CA') | (flights_FL_201407['DEST_CITY_NAME'] == 'New York, NY'))]
MIA_NYC_LA_201307 = flights_FL_201307[(flights_FL_201307['ORIGIN_CITY_NAME'] == 'Miami, FL') & ((flights_FL_201307['DEST_CITY_NAME'] == 'Los Angeles, CA') | (flights_FL_201307['DEST_CITY_NAME'] == 'New York, NY'))]
```

```python
print(MIA_NYC_LA_201807.shape)
print(MIA_NYC_LA_201707.shape)
print(MIA_NYC_LA_201607.shape)
print(MIA_NYC_LA_201507.shape)
print(MIA_NYC_LA_201407.shape)
print(MIA_NYC_LA_201307.shape)
```

```python
miami_07flights = pd.concat([MIA_NYC_LA_201307, MIA_NYC_LA_201407,
                             MIA_NYC_LA_201507, MIA_NYC_LA_201607,
                             MIA_NYC_LA_201707, MIA_NYC_LA_201807]).reset_index(drop = True)
```

```python
miami_07flights.describe()
```

```python
print(sum(miami_07flights['DEP_TIME'].isna()))
print(sum(miami_07flights['CRS_DEP_TIME'].isna()))
print(sum(miami_07flights['ARR_TIME'].isna()))
print(sum(miami_07flights['CRS_ARR_TIME'].isna()))
miami_07flights_droptimena = miami_07flights.dropna(subset = ['DEP_TIME', 'CRS_DEP_TIME', 
                                                              'ARR_TIME', 'CRS_ARR_TIME'])
```

```python
miami_07flights_test = miami_07flights_droptimena.replace({np.nan : None})
```

```python
# keep columns of interest for flight data
miami_07flights_subset = miami_07flights_test[['FL_DATE', 'ORIGIN', 'ORIGIN_CITY_NAME', 
                                                     'DEST', 'DEST_CITY_NAME', 'CRS_DEP_TIME', 
                                                     'DEP_TIME', 'DEP_DELAY', 'TAXI_OUT', 'WHEELS_OFF', 
                                                     'CRS_ARR_TIME', 'ARR_TIME', 'ARR_DELAY', 'TAXI_IN', 
                                                     'WHEELS_ON', 'CANCELLED', 'CANCELLATION_CODE', 
                                                     'DIVERTED', 'CRS_ELAPSED_TIME', 'ACTUAL_ELAPSED_TIME', 
                                                     'AIR_TIME', 'CARRIER_DELAY', 'WEATHER_DELAY', 
                                                     'NAS_DELAY', 'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY']]
miami_07flights_tuples = [tuple(row) for row in miami_07flights_subset.values]
```

```python
miami_07flights_tuples
```

```python
miami_07flights_subset
```

```python
miami_07flights_droptimena.to_csv('miami_07flights.csv')
```
